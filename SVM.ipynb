```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target'], axis='columns')
```

```python
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test, y_test)
```
